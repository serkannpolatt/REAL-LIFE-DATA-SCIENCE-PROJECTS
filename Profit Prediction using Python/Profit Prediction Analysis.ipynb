# Profit Prediction using Python
The dataset that I am using for the task of profit prediction includes data about the R&D spend, Administration cost, Marketing Spend, State of operation, and the historical profit generated by 50 startups. 

# Python kullanarak Kar Tahmini
Kar tahmini görevi için kullandığım veri kümesi, Ar-Ge harcaması, Yönetim maliyeti, Pazarlama Harcaması, Operasyon Durumu ve 50 yeni girişimin oluşturduğu geçmiş kâr hakkındaki verileri içerir.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Startups.csv")
```

```python
data.head()
```

### This data doesn’t contain any missing values so without wasting any time let’s start by having a look at the summary statistics of this data:

### Bu veriler herhangi bir eksik değer içermiyor, bu yüzden vakit kaybetmeden bu verilerin özet istatistiklerine bir göz atarak başlayalım:

```python
data.describe()
```

### Now let’s have a look at the correlation between the features:

### Şimdi özellikler arasındaki ilişkiye bir göz atalım:

```python
plt.figure(figsize=(12,10))
sns.heatmap(data.corr(),annot=True)
```

### As this task is based on the problem of regression so I will be using the Linear regression algorithm to train the profit prediction model. So let’s prepare the data so that we can fit it into the model:

### Bu görev regresyon problemine dayandığından, kar tahmin modelini eğitmek için Lineer regresyon algoritmasını kullanacağım. Öyleyse verileri modele sığdırabilmemiz için hazırlayalım:

```python
x=data[["R&D Spend","Administration","Marketing Spend"]]
```

```python
y=data["Profit"]
```

```python
x=x.to_numpy()
y=y.to_numpy()
```

```python
y=y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

### Now let’s train a linear regression model on this data and have a look at the predicted values:

### Şimdi bu veriler üzerinde bir lineer regresyon modeli eğitelim ve tahmin edilen değerlere bir göz atalım:

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
data = pd.DataFrame(data={"Predicted Profit": ypred.flatten()})
```
