```python
def bar_plot(variable):
    var = data[variable]
    varValue = var.value_counts()
    
    plt.figure(figsize = (9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel('Frequency')
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
categories = ['Cinsiyet','Yas','Bolge','Egitim','parti']
for c in categories:
    bar_plot(c)
```

```python
# Mapping Cinsiyet
data.Cinsiyet=[1 if each=='Erkek' else 0 for each in data.Cinsiyet]
```

```python
# Mapping questions
data.EKONOMIK_DURUMUM_IYI=[1 if each=='Evet' else 0 for each in data.EKONOMIK_DURUMUM_IYI]
data.EGITIME_REFORM=[1 if each=='Evet' else 0 for each in data.EGITIME_REFORM]
data.OZELLESTIRME_YANLISI=[1 if each=='Evet' else 0 for each in data.OZELLESTIRME_YANLISI]
data.IDAM_CEZASI=[1 if each=='Evet' else 0 for each in data.IDAM_CEZASI]
data.HUR_BASIN_VARMI=[1 if each=='Evet' else 0 for each in data.HUR_BASIN_VARMI]
data.ALKOL_YASAGINI_DESTEKLIYOR=[1 if each=='Evet' else 0 for each in data.ALKOL_YASAGINI_DESTEKLIYOR]
data.LAIK_DEVLET_ISTEGI=[1 if each=='Evet' else 0 for each in data.LAIK_DEVLET_ISTEGI]
data.KURTAJ_YASASINA_DESTEK=[1 if each=='Evet' else 0 for each in data.KURTAJ_YASASINA_DESTEK]
data.OHAL_OZGURLUGU_KISITLIYORMU=[1 if each=='Evet' else 0 for each in data.OHAL_OZGURLUGU_KISITLIYORMU]
data.YENI_PARTILERE_DESTEK=[1 if each=='Evet' else 0 for each in data.YENI_PARTILERE_DESTEK]
```

```python
# Mapping Education
educational_mapping= {'Lisans Üstü':5,"Lisans":4,"Ön Lisans":3,"Lise":2,'Ortaokul':1,"İlkokul":0}
data.Egitim = data.Egitim.map(educational_mapping)
```

```python
# Mapping Political Choices
parti_mapping = {   'IYI PARTI':5,
                    "CHP":4,
                    "DIĞER":3,
                    "AKP":2,
                    "MHP":1,
                    "HDP":0}
data.parti = data.parti.map(parti_mapping)
```

```python
# Mapping Ages
age_mapping = {'0-18':0,'18-30':1,'30-50':2,'50-60':3,'60+':4}
data.Yas = data.Yas.map(age_mapping)
```

```python
# Mapping Bolge
bolge_mapping = {'Marmara':0,'Ege':1,'İç Anadolu':2,'Akdeniz':3,'Karadeniz':4,'Güneydoğu':5,'Doğu Anadolu':6}
data.Bolge = data.Bolge.map(bolge_mapping)
data.head(10)
```

```python
x = data.drop(['parti'],axis=1)
y = data.parti

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.33, random_state = 42)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
acc_log_train = round(logreg.score(x_train,y_train)*100,2)
acc_log_test = round(logreg.score(x_test,y_test)*100,2)
print("Training Accuracy: % {}".format(acc_log_train))
print("Testing Accuracy: % {}".format(acc_log_test))
```
