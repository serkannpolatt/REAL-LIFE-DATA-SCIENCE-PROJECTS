# Predict whether the Employee of an Organization should get Promotion or Not?
# Bir Kuruluşun Çalışanının Terfi Alıp Almayacağını Tahmin Edin?

Your client is a large MNC and they have 15 broad verticals across the organisation. One of the problem your client is facing is around identifying the right people for promotion (only for manager position and below) and prepare them in time. Currently the process, they are following is: * They first identify a set of employees based on recommendations/ past performance. * Selected employees go through the separate training and evaluation program for each vertical. These programs are based on he required skill of each vertical * At the end of the program, based on various factors such as training performance, an employee gets the promotion
### -----------------------------------------------------
Müşteriniz büyük bir ÇUŞ'tur ve kuruluş genelinde 15 geniş sektöre sahiptir. Müşterinizin karşılaştığı sorunlardan biri, terfi için doğru kişileri belirlemek (sadece yönetici pozisyonu ve altı için) ve onları zamanında hazırlamaktır. Şu anda takip ettikleri süreç şu şekildedir: * Önce tavsiyelere/geçmiş performansa dayalı olarak bir dizi çalışanı belirlerler. * Seçilen çalışanlar her sektör için ayrı eğitim ve değerlendirme programından geçmektedir. Bu programlar, her dikeyin gerektirdiği beceriye dayalıdır * Programın sonunda, eğitim performansı gibi çeşitli faktörlere bağlı olarak, bir çalışan terfi alır.

```python
# handling Imblanced data

from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X, y = oversample.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split( X, y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_valid.shape)
print(y_valid.shape)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier

sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_valid_scaled = sc.transform(x_valid)
x_test_scaled = sc.transform(x_test)

model = DecisionTreeClassifier()
model.fit(x_train_scaled, y_train)

y_train_pred = model.predict(x_train_scaled)
y_valid_pred = model.predict(x_valid_scaled)

print("Training Acc/eğitim doğruluğu.: ", model.score(x_train_scaled, y_train))
print("Validation Acc/doğrulama doğruluğu.: ", model.score(x_valid_scaled, y_valid))  
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_valid, y_valid_pred)  # Use y_valid_pred instead of y_train_pred
sns.heatmap(cm, annot=True, cmap='Wistia')
plt.xlabel("Predicted values")
plt.ylabel("Actual values")
plt.show()
```

```python
cr = classification_report(y_valid, y_valid_pred)
print(cr)
```
