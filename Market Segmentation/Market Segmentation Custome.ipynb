### Scaling the DataFrame / DataFrame'i ölçeklendirme

```python
scaled_df = scalar.fit_transform(df)
```

### Dimensionality reduction / Boyutsal küçülme
Converting the DataFrame into 2D DataFrame for visualization

DataFrame'i görselleştirme için 2D DataFrame'e dönüştürme

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)
pca_df = pd.DataFrame(data=principal_components ,columns=["PCA1","PCA2"])
pca_df
```

### Hyperparameter tuning / Hiperparametre ayarı
Finding 'k' value by Elbow Method

'k' değerini Dirsek Yöntemi ile bulma

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K / Knın değerleri') 
plt.ylabel('Inertia / Eylemsizlik') 
plt.title('The Elbow Method using Inertia / Atalet Kullanan Dirsek Yöntemi') 
plt.show()
```

###  Model Building using KMeans / KMeans Kullanarak Model Oluşturma

```python
kmeans_model=KMeans(4)
kmeans_model.fit_predict(scaled_df)
pca_df_kmeans= pd.concat([pca_df,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

### Visualizing the clustered dataframe / Kümelenmiş veri çerçevesini görselleştirme 

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x="PCA1",y="PCA2",hue="cluster",data=pca_df_kmeans,palette=['red','green','blue','black'])
plt.title("Clustering using K-Means Algorithm / K-Ortalamalar Algoritmasını Kullanarak Kümeleme")
plt.show()
```

```python
# find all cluster centers
# tüm küme merkezlerini bul
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
# inverse transform the data
# verileri ters dönüştürme
cluster_centers = scalar.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
# Creating a target column "Cluster" for storing the cluster segment
# Küme segmentini depolamak için bir hedef sütun "Küme" oluşturma
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_df
```

# Training and Testing the model accuracy using decision tree
# Karar ağacını kullanarak model doğruluğunun eğitimi ve test edilmesi

```python
# Split Dataset
# Veri Kümesini Böl
X = cluster_df.drop(['Cluster'],axis=1)
y= cluster_df[['Cluster']]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
X_test
```

```python
# Decision_Tree
# Karar ağacı
model= DecisionTreeClassifier(criterion="entropy")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# Confusion_Matrix
# Karışıklık_Matrix
print(metrics.confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Saving the Decision tree model for future prediction / Gelecek tahmini için Karar ağacı modelini kaydetme

```python
import pickle
filename = 'final_model.sav'
pickle.dump(model, open(filename, 'wb'))
 
 
# load the model from disk
# modeli diskten yükle
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result,'% Acuuracy')
```
