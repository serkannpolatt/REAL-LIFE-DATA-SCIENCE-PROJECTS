```python
def get_address(row):
  address = ""

  nerd_xd_2 = row
  for n_idx, n in enumerate(nerd_xd_2):
    print(n_idx, n)

    if n_idx !=0:
      # print(n_idx)
      if n["start"] == nerd_xd_2[n_idx-1]["end"]:
        address = address + n["word"]
      else:
        address = address + " " + n["word"]
    else:
      address = n["word"]
  return address
```

```python
df["parsed_address"] = df["parsed"].apply(lambda x: get_address(x))
```

```python
def transformers_inference(model_id):
  
  pipe = pipeline("token-classification", model_id)
  inference_list = []
  for i in range(30):
    inference_list.append({"input": dataset["hits"]["hits"][i]["_source"]["text"],
                    "parsed": pipe(dataset["hits"]["hits"][i]["_source"]["text"])})
  
  df = pd.DataFrame.from_records(inference_list)
  df.head(10)

  df["parsed_address"] = df["parsed"].apply(lambda x: get_address(x))
  return df
```

```python
deprem_ner_df = transformers_inference("deprem-ml/deprem-ner")
```
